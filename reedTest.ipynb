```python
from bs4 import BeautifulSoup
import requests
import json
import pickle
import queue
import threading
import time
import operator
```

```python
global q, count
count = 0
q = queue.Queue()
threadAlive = 0
with open("links.pickle", "rb") as f:
    links = pickle.load(f)
```

```python
print(len(links))
```

```python
def scrape(links):
    global count, q, threadAlive
    for link in links:
        try:
            #print(link)
            page_response = requests.get(link, timeout=5)
            page_content = BeautifulSoup(page_response.content, "html.parser")
            enquired = 0
            purchased = 0
            awardingBody = "None"
            #script = page_content.find('script')
            #print(page_content.prettify())
            tmpData = dict()

            try:
                soup = page_content.find("section", {"class": "border-bottom-0"})
                dtemp = soup.text.strip()
                if "No formal qualification" in dtemp:
                    awardingBody = "None"
                elif "Professional" in dtemp and "CPD" in dtemp:
                    awardingBody = "Professional, CPD"
                elif "Professional" in dtemp:
                    awardingBody = "Professional"
                elif "CPD" in dtemp:
                    awardingBody = "CPD"
                else:
                    pass

            except Exception as e:
                print(e, " ---- Error in scrape function")


            try:
                for div in page_content.findAll('div', {'id': 'number-enquiries-purchases'}):
                    saleOrEnqInfo = div.text.strip()
                    break
                if "enquired" in saleOrEnqInfo:
                    saleOrEnqInfo = saleOrEnqInfo.split(" ")
                    enquired = saleOrEnqInfo[0]
                if "purchased" in saleOrEnqInfo:
                    saleOrEnqInfo = saleOrEnqInfo.split(" ")
                    purchased = saleOrEnqInfo[0]
            except:
                pass

            #print(awardingBody, enquired, purchased)


            for tag in page_content.findAll('script'):
                # Use extract to remove the tag
                if "dataLayer.push" in str(tag.extract()) and "courseID" in str(tag.extract()):
                    temp = str(tag.extract())
                    temp = temp[temp.find("(")+1:temp.find(");")]
                    #print(temp)
                    temp = temp.replace("'",'"')
                    temp1 = json.loads(temp)
                    #print(temp1)
            tmpData = {
                    "courseLink": link,
                    "courseName": temp1['courseName'],
                    "courseSubject": temp1['courseSubject'],
                    "coursePrice": temp1['coursePrice'],
                    "courseProvider": temp1['courseProvider'],
                    "Purchase": purchased,
                    "Enquiry": enquired,
                    "courseAwardingBody": awardingBody
                }
            #print(tmpData)   
            count+=1
            q.put(tmpData.copy())
            
        except Exception as e:
            print(e, " ---- Error in scrape function")
    threadAlive-=1
    
```

```python
links = list(links)
threading.Thread(target = scrape, args=[links[:920], ]).start()
threading.Thread(target = scrape, args=[links[920:1840], ]).start()
threading.Thread(target = scrape, args=[links[1840:2760], ]).start()
threading.Thread(target = scrape, args=[links[2760:3680], ]).start()
threading.Thread(target = scrape, args=[links[3680:4601], ]).start()
threadAlive = 5
time.sleep(5)
def main():
    data = list()
    while True:
        global count, q, threadAlive
        while not q.empty():
            temp = q.get()
            data.append(temp.copy())
            print("Count - ", count, threadAlive)
        if threadAlive<1:
            break
        
    try:
        with open("dataNew.pickle", "wb") as f:
            pickle.dump(data, f)
    except Exception as e:
            print(e) 
main()
```
