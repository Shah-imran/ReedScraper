```python
from bs4 import BeautifulSoup
import requests
```

```python
links = list()
for i in range(1,185,1):
    page_link = "https://www.reed.co.uk/courses/writing"
    if i != 1:
        page_link = "https://www.reed.co.uk/courses/writing" + "?pageno=" + str(i)
        
    print(page_link)
    print("Page No - ", i)
    page_response = requests.get(page_link, timeout=5)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    #print(page_content.prettify())
    for a in page_content.find_all('a', href=True):
        if "#/courses/writing" in a['href']:
            links.append("https://www.reed.co.uk"+a['href'])
links = set(links)
#print(links)
print(len(links))
```

```python
import pickle
with open("links.pickle", "wb") as f:
    pickle.dump(links, f)
```

```python
from slimit import ast  # $ pip install slimit
from slimit.parser import Parser as JavascriptParser
from slimit.visitors import nodevisitor
import json
```

```python
data = list()
count = 0
for link in links:
    try:
        #link = "https://www.reed.co.uk/courses/copywriting-course-level-4-diploma/57653#/courses/writing"
        print("Downloading - ", count)
        print(link)
        page_response = requests.get(link, timeout=5)
        page_content = BeautifulSoup(page_response.content, "html.parser")
        enquired = 0
        purchased = 0
        #script = page_content.find('script')
        #print(page_content.prettify())
        tmpData = dict()
        try:
            for div in page_content.findAll('div', {'id': 'number-enquiries-purchases'}):
                saleOrEnqInfo = div.text.strip()
                break
            if "enquired" in saleOrEnqInfo:
                saleOrEnqInfo = saleOrEnqInfo.split(" ")
                enquired = saleOrEnqInfo[0]
            if "purchased" in saleOrEnqInfo:
                saleOrEnqInfo = saleOrEnqInfo.split(" ")
                purchased = saleOrEnqInfo[0]
        except:
            pass

        print(enquired, purchased)


        for tag in page_content.findAll('script'):
            # Use extract to remove the tag
            if "dataLayer.push" in str(tag.extract()) and "courseID" in str(tag.extract()):
                temp = str(tag.extract())
                temp = temp[temp.find("(")+1:temp.find(");")]
                #print(temp)
                temp = temp.replace("'",'"')
                temp1 = json.loads(temp)
                #print(temp1)
        tmpData = {
                "courseLink": link,
                "courseName": temp1['courseName'],
                "courseSubject": temp1['courseSubject'],
                "coursePrice": temp1['coursePrice'],
                "courseProvider": temp1['courseProvider'],
                "Purchase": purchased,
                "Enquiry": enquired
            }
        data.append(tmpData.copy())
        #print(tmpData)
        count+=1    
    except Exception as e:
        print(e)
        
print(len(data))
try:
    with open("data.pickle", "wb") as f:
        pickle.dump(data, f)
except Exception as e:
    print(e)
```
